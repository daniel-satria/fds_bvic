```python
df_rolled = df.join(
    df,
    left_on="from_account_number", 
    right_on="from_account_number", 
    how="left",
    suffix="_r").filter(
        (pl.col("transaction_date_r") >= pl.col("transaction_date") - pl.duration(hours=24)) & 
        (pl.col("transaction_date_r") <= pl.col("transaction_date"))
) 

# Step 3: Group by original row (using unique identifier) to sum amounts within 24h 
df_sums = df_rolled.group_by("no_referensi").agg(
    pl.sum("transaction_amount_r").alias("rolling_24h_sum")
) 

# Step 4: Merge rolling sums back to original DataFrame 
df_flagged = df.join(
    df_sums, 
    on="no_referensi"
    ).with_columns([
        (pl.col("rolling_24h_sum") >= 50_000_000).alias("flag_50M")]).sort(["from_account_number", "transaction_date"])
```

```python
# Step 1: Self-join on account number within 24h window
df_rolled = df.join(
    df,
    left_on="from_account_number",
    right_on="from_account_number",
    how="left",
    suffix="_r"
).filter(
    (pl.col("transaction_date_r") >= pl.col("transaction_date") - pl.duration(hours=24)) &
    (pl.col("transaction_date_r") <= pl.col("transaction_date"))
)

# Step 2: Group by original row and sum amounts
df_sums = df_rolled.group_by("no_referensi").agg(
    pl.sum("transaction_amount_r").alias("rolling_24h_sum")
)

# Step 3: Identify which accounts exceeded 50M in 24h
df_flagged_accounts = df_rolled.join(df_sums, on="no_referensi").filter(
    pl.col("rolling_24h_sum") >= 50_000_000
)

# Step 4: Extract contributing transaction IDs
flagged_refs_list = df_flagged_accounts.select("no_referensi_r").collect().to_series().to_list()

# Step 5: Merge rolling sum and flag back to original DataFrame
df_flagged = df.join(df_sums, on="no_referensi").with_columns([
    pl.col("no_referensi").is_in(flagged_refs_list).alias("flag_50m")
]).sort(["from_account_number", "transaction_date"])
```
