```python
# Step 1: Self-join on account number within 24h window
df_rolled = (
    df.join(
        df,
        on="from_account_number",   # same as left_on/right_on
        how="left",
        suffix="_r"
    )
    .filter(
        (pl.col("transaction_date_r") >= pl.col("transaction_date") - pl.duration(hours=24)) &
        (pl.col("transaction_date_r") <= pl.col("transaction_date"))
    )
)

# Step 2: Group sums on the original transaction
df_sums = (
    df_rolled
    .group_by("no_referensi")
    .agg(pl.sum("transaction_amount_r").alias("rolling_24h_sum"))
)

# Step 3: Join sums back and filter >= 50M
df_flagged_accounts = (
    df_rolled
    .join(df_sums, on="no_referensi")
    .filter(pl.col("rolling_24h_sum") >= 50_000_000)
)

# Step 4: EXTRACT KEYS (these transactions exceeded 50M)
keys = (
    df_flagged_accounts
    .select(["from_account_number", "no_referensi"])
    .unique()
)

# Step 5: Returns only rows that match the keys
flagged = (
    df.join(keys, on=["from_account_number", "no_referensi"], how="semi")
      .with_columns(pl.lit(1).alias("flag"))
)

# Bring the flag back into df
# df_b = (
#     df.join(flagged, on=["from_account_number", "no_referensi"], how="left")
#       .with_columns(
#           pl.col("flag").fill_null(pl.col("flag_50mio"))
#       )
# )

df_b = (
    df.join(
        keys,
        on=["from_account_number", "no_referensi"],
        how="left",
        suffix="_key"
    )
    .with_columns(
        pl.when(pl.col("from_account_number").is_not_null())
        .then(1)
        .otherwise(pl.col("flag_50mio"))
        .alias("flag_50mio")
    )
    .drop(["from_account_number", "no_referensi"])
)
```
